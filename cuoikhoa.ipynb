```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import plotly.express as px
```

```python
df = pd.read_csv("cardiovascular-disease-death-rate-who-mdb.csv")
df
```

```python
df_venezuela = df.loc[df['Entity']=='Venezuela', :]
df_venezuela
```

```python
# List of columns with missing values
columns_with_missing = df.columns[df.isnull().any()]

# Impute numerical columns with mean
numerical_columns = df.select_dtypes(include=['float64'])
numerical_columns = numerical_columns.columns[numerical_columns.isnull().any()]
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
```

```python
x= df_venezuela["Year"]
y= df_venezuela["Age-standardized deaths from cardiovascular diseases in both sexes in those aged all ages per 100,000 people"]
```

```python
plt.figure(figsize=(80, 40))
plt.plot(x, y, marker='o', markersize=20, linestyle="-", linewidth=5)

plt.xlabel('Year - Entity', fontsize=50)
plt.ylabel('People died', fontsize=50)
plt.title("Age-standardized deaths from cardiovascular diseases in both sexes in those aged all ages per 100,000 people in Venezuela", fontsize=50)
```
